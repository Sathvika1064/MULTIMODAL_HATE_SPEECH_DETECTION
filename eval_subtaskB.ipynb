```
for i in range(len(fnames)):
  path.append(fnames[i])
  filepath=os.path.join(root,fnames[i])
  output = reader.readtext(filepath)
  for x in range(len(output)):
    arr.append(output[x][1])
  result.append(arr)
  arr=[]

for x in result:
  ans = ' '
  for i in x:

  # concatenating the strings
  # using + operator
    ans = ans+ ' '+ i
  r.append(ans)
  ans=''
```
